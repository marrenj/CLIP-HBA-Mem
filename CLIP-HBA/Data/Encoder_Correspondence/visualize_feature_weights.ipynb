```python
import numpy as np

path = "./weighting_matrix/weighting_matrix_things_cv.npy"

weights = np.load(path)
print(weights.shape)
np.round(weights, 6)

# plot

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5, 8))
sns.heatmap(weights, cmap="viridis", cbar=False)
y_axis = list(range(-100, 1301, 5))
plt.yticks(range(0, 281, 10), y_axis[::10])
plt.ylabel("Time (ms)")
plt.xlabel("Layer")
plt.title("Weighting matrix")
plt.tight_layout()
plt.show()
```
