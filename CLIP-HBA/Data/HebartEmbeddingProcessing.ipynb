# Target Embedding Processing

```python
import pandas as pd
import numpy as np

path = "./spose_embedding_66d_sorted.txt"
df = pd.DataFrame(np.loadtxt(path))
df = df.dropna(axis=1)
df_min = df.min().min()
df_max = df.max().max()
df = ((df - df_min) / (df_max - df_min))*100
df['image_name'] = pd.read_csv("./hebart66_embedding_rescaled.csv").iloc[:, 0]
df = df[['image_name'] + [col for col in df.columns if col != 'image_name']]
df
```

```python
df.to_csv("hebart66_embedding_rescaled100.csv", index=True)
```

Drop the 48 Validation Sample set

```python
from scipy.io import loadmat
sample48 = pd.DataFrame(loadmat("./words48.mat")['words48'])
sample48 = sample48.astype(str).apply(lambda x: x.str.strip('[]'))
sample48 = sample48.applymap(lambda x: x[1:-1])
sample48 = sample48.replace(' ', '_', regex=True)
sample48 = sample48.replace('camera', 'camera1', regex=True)
sample48 = sample48.replace('file', 'file1', regex=True)

print(np.array(sample48).tolist())
```

```python
df_train = df[~df['image_name'].str[:-8].isin(sample48[0].values.tolist())]
df_train
```

```python
df_train.to_csv("hebart66_embedding_rescaled100_1806train.csv", index=True)
```

```python
df_val = df[df['image_name'].str[:-8].isin(sample48[0].values.tolist())]
df_val
```

```python
df_val.to_csv("hebart66_embedding_rescaled100_48val.csv", index=True)
```
